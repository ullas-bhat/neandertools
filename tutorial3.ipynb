# Pipeline run

```python
from neandertools import AsteroidCutoutPipeline
from neandertools.imagebuilder import cutouts_grid
```

```python
pipeline = AsteroidCutoutPipeline(
      target="2024 TN57",
      target_type="smallbody",
      start="2024-12-01",
      end="2024-12-30",
      dr="dp1",
      collection="LSSTComCam/DP1",
      bands=["u", "g", "r", "i", "z", "y"],   
      step="12h",                   
      cutout_size=150,             
      polygon_widening_arcsec=2, 
  )

gif_path = pipeline.run(
    output_path="asteroid3.gif",
    match_background=True,
    match_noise=True,
  )
```

```python
fig, axes = cutouts_grid(pipeline.cutouts, metadata=pipeline.frame_metadata, match_noise=True)
```
