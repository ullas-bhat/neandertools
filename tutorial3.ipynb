# Pipeline run

```python
from neandertools import AsteroidCutoutPipeline

pipeline = AsteroidCutoutPipeline(                                                                                                                                      
      target="2024 TN57",
      target_type="smallbody",
      start="2024-12-01",
      end="2024-12-30",
      dr="dp1",
      collection="LSSTComCam/DP1",
      bands=["u", "g", "r", "i", "z", "y"],   
      step="12h",                   
      cutout_size=150,             
      polygon_widening_arcsec=2,
  )
```

```python
gif_path = pipeline.run(
      output_path="asteroid.gif",
      frame_duration_ms=500,
      match_background=True,
      show_ne=True,
      warp_common_grid=True,
  )
```

```python
fig, axes = pipeline.grid(ncols=5, show_ne=True, warp_common_grid=True)
```
